```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Automobile price data _Raw_.csv')
pd.set_option('display.max_columns', None)
# df.head()
```

```python
df.info()
```

# <br><br> <font color=purple>**Replacing '?' with NaN and then filled missing values by mean** </font><br><br>

```python
df = df.replace('?',np.NAN) 
df.isnull().sum()
```

```python
a=df[df['normalized-losses']!='?']
b=(a['normalized-losses'].astype(float)).mean()
df['normalized-losses']=df['normalized-losses'].fillna(b)
```

```python
a=df['num-of-doors'].map({'two':2,'four':4,'NaN':4})
df['num-of-doors']=a
df['num-of-doors'] = df['num-of-doors'].fillna(4)
```

```python
a=df[df['price']!='?']
b=(a['price'].astype(float)).mean()
df['price']=df['price'].fillna(b)
```

```python
a=df[df['horsepower']!='?']
b=(a['horsepower'].astype(float)).mean()
df['horsepower']=df['horsepower'].fillna(b)
```

```python
a=df[df['bore']!='?']
b=(a['bore'].astype(float)).mean()
df['bore']=df['bore'].fillna(b)
```

```python
a=df[df['stroke']!='?']
b=(a['stroke'].astype(float)).mean()
df['stroke']=df['stroke'].fillna(b)
```

```python
a=df[df['peak-rpm']!='?']
b=(a['peak-rpm'].astype(float)).mean()
df['peak-rpm']=df['peak-rpm'].fillna(b)
```

```python
a=df['num-of-cylinders'].map({'four':4,'five':5,'six':6,'nan':4})
df['num-of-cylinders']=a
df['num-of-cylinders']=df['num-of-cylinders'].fillna(4)
```

# <br><br><font color=darkred>Categorical feature treatment One Hot Encoding <font><br><br>

```python
object_col = []
for i in df.columns:
    if df[i].dtype == object:
        object_col.append(i)
```

```python
print(object_col)
```

```python
from sklearn.preprocessing import OneHotEncoder
enc = OneHotEncoder(handle_unknown='ignore')
```

```python
enc_df = pd.DataFrame(enc.fit_transform(df[object_col]).toarray())
```

```python
df = df.join(enc_df)
df.shape
```

```python
df.drop(object_col, axis=1,inplace=True)
```

```python
# df.info()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop(["price"],axis=1)
y = df["price"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,y,train_size=0.7,random_state=42)
```

```python
print("Size of x train is",x_train.shape)
print("Size of y train is",y_train.shape)
print("Size of x test is",x_test.shape)
print("Size of y test is",y_test.shape)
```

```python

from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(x_train, y_train)
```

```python

print('Coefficients:' ,lin_reg.coef_)
print('Intercept:' ,lin_reg.intercept_)
print('R2 score' ,lin_reg.score(X, y))
# print('MSE,MAE',MSE(y_train,))
```

```python
import statsmodels.api as sm
X_constant = sm.add_constant(X)
lin_reg = sm.OLS(y,X_constant).fit()
lin_reg.summary()
```

```python
y_pred = lin_reg.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error as mae

mae(y_test,y_pred)
```
